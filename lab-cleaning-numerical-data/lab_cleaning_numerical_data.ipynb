```python
# 3. First look at its main features (head, shape, info).

def data_exploration (dataframe):
    # Display the shape of the DataFrame
    print("DataFrame Shape:")
    print(dataframe.shape)
    print("\n")

    # Display the summary statistics of the DataFrame
    print("DataFrame Summary Statistics:")
    print(dataframe.describe())
    print("\n")

    # Display the data types of each column
    print("Data Types of Columns:")
    print(dataframe.dtypes)
    print("\n")

    # Display the count of missing values for each column
    print("Missing Values Count per Column:")
    print(dataframe.isna().sum())
```

```python
#Call the function to check for data shape data info, data types and null values
data_exploration (df)
```

```python
# 4. Rename the columns so they follow the PE8 (snake case) (lower_case_with_underscores)
#funtion for cleaning data 
 
def clean_data (dataframe):
    cols = [] 

    for i in range(len(dataframe.columns)):
        cols.append(dataframe.columns[i].lower().replace(' ','_'))
    dataframe.columns = cols

    return dataframe
```

```python
# call the funtion to standerdise the data frame
clean_data (df)
```

```python
# 7. Define a function that differentiates between continuous and discrete variables. Hint: The number of unique values might be useful. Store continuous data into a continuous_df variable and do the same for discrete_df

def diff_variables(data_frame):
    
    global continuous_df
    global discrete_df
    
    continuous_df = data_frame
    discrete_df = data_frame
    
    for col in data_frame.columns:
        if data_frame[col].dtype == "float":
            discrete_df = discrete_df.drop([col], axis = 1)
            print(col, "is a continuous variable")
        else:
            continuous_df = continuous_df.drop([col], axis = 1)
            print(col, "is a discrete variable")
```

```python
#Call the function
diff_variables(numerical_data)
continuous_df
discrete_df
```

```python
# 9. Create a function to plot every discrete variable. Do the same with the continuous variables (Be Careful, you may need to change the plot type to one better suited for continuous data!)
#function for continous variables 
def plot_continuous(dataframe):
    
    for col in dataframe.columns:
        sns.distplot(dataframe[col])
        plt.show()                

def plot_discrete(dataframe):
    
    for col in dataframe.columns:
        sns.histplot(data = dataframe, x = col, binwidth = dataframe[col].max()/20)
        plt.show()
        
```
