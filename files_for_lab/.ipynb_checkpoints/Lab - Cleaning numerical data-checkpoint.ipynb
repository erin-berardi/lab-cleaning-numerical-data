```python
# 6. Create a new dataframe using only the numerical data called numerical_df. Use this for the following questions.
numericals_df = customer_df.select_dtypes([np.float64,np.int64])
numericals_df
```

```python
# 7. Define a function that differentiates between continuous and discrete variables. Hint: The number of unique values 
# might be useful. Store continuous data into a continuous_df variable and do the same for discrete_df
numericals_df.dtypes
```

```python
for col in numericals_df.columns:
    uniques = len(numericals_df[col].value_counts())
    print(col,': ',uniques)
```

```python
def differentiate(dataframe):
    continuous_cols = []
    discrete_cols = []
    limit = 20
    for col in dataframe.columns:
        if len(dataframe[col].value_counts()) <= limit:
            discrete_cols.append(col)
        else:
            continuous_cols.append(col)
    continuous_df = dataframe[continuous_cols]
    discrete_df = dataframe[discrete_cols]
    
    return continuous_df, discrete_df
continuous_df, discrete_df = differentiate(numericals_df)
```

```python
# 9. Create a function to plot every discrete variable. Do the same with the continuous variables (Be Careful, you may need to 
# change the plot type to one better suited for continuous data!)

def discrete_plot(DataFrame):
    for col in DataFrame.columns:
        DataFrame[col].value_counts().plot(kind='bar')
        plt.title(col)
        plt.show()
    return
discrete_plot(discrete_df)
```

```python
def continuous_plot(DataFrame):
    for col in DataFrame.columns:
        sns.histplot(x=DataFrame[col])
        plt.title(col)
        plt.show()
    return
continuous_plot(continuous_df)
```

```python
# 10. Comment what you can see in the plots.

# Both discrete data variables have a left skewed distribution and a small number of unique values.
# Most continuous data variables are left skewed and the total_claim_amount has a very high number of uniques values. But it is
# also the most normal-shaped distribution
```

```python
# 11. Look for outliers in the continuous variables. (HINT: Thereâ€™s a good plot to do that!)
def continuous_boxplot(DataFrame):
    for col in DataFrame.columns:
        sns.boxplot(x=DataFrame[col])
        plt.title(col)
        plt.show()
    return
continuous_boxplot(continuous_df)
```

```python
# 12. Did you find outliers? Comment what you will do with them. If you decide to remove them, do that now.

# I found several outliers and will replace them by the mean.

for col in continuous_df.columns:
    iqr = np.percentile(continuous_df[col],75) - np.percentile(continuous_df[col],25)
    upper_limit = np.percentile(continuous_df[col],75) + 1.5*iqr
    lower_limit = np.percentile(continuous_df[col],25) - 1.5*iqr
    mean = continuous_df[col].mean()
    continuous_df.loc[continuous_df[col]>upper_limit, col] = mean
    continuous_df.loc[continuous_df[col]<lower_limit, col] = mean
```
