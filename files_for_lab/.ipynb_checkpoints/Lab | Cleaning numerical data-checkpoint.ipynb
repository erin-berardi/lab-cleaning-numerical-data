```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv' ,index_col=0)
data.head()
```

```python
new_column_names = {}
for col in data.columns:
    new_col = col.lower().replace(' ', '_')
    new_column_names[col] = new_col

data = data.rename(columns=new_column_names)
```

```python
data.columns
```

```python
data['effective_to_date'] = pd.to_datetime(data['effective_to_date'],errors='coerce')
```

```python
data.head()
```

```python
data.dtypes
```

```python
numerical_df = data.select_dtypes(include = np.number) 
```

```python
numerical_df.dtypes
```

```python
for col in numerical_df.columns:
    uniques = len(numerical_df[col].value_counts())
    print(col,': ',uniques)
```

```python
def differentiate_variables(df):
    continuous_cols = []
    discrete_cols = []

    for col in numerical_df.columns:
        if numerical_df[col].nunique() <= 20:
            discrete_cols.append(col)
        else:
            continuous_cols.append(col)

    continuous_df = df[continuous_cols]
    discrete_df = df[discrete_cols]

    return continuous_df, discrete_df
continuous_df, discrete_df = differentiate_variables(numerical_df)


```

```python
continuous_df
```

```python
discrete_df
```

```python
corr_matrix = numerical_df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()


```

```python
def plot_discrete_variables(df):
    for var in discrete_df:
        plt.figure()
        discrete_df[var].value_counts().plot(kind='bar')
        plt.title(var)
        plt.show()
plot_discrete_variables(discrete_df)
```

```python
def plot_continuous_variables(df):
    for col in df.columns:
        plt.figure()
        sns.boxplot(x=df[col])
        plt.title(col)
        plt.show()
plot_continuous_variables(continuous_df)
```

```python
def plot_continuous_variables(df):
    for col in df.columns:
        plt.figure(figsize=(8, 6))
        sns.histplot(df[col], bins=22, kde=True)
        plt.title(col + ' Histogram')

        plt.tight_layout()
        plt.show()
plot_continuous_variables(continuous_df)
```
