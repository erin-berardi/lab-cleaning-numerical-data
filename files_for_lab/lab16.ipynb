```python
def variable_type(column, numerical_df):
    column_values = numerical_df[column]
    column_range = column_values.max() - column_values.min()
    if column_range < 10:
        return "discrete"
    else:
        return "continuous"
```

```python
def plot_discrete_vars(numerical_df):
    for col in numerical_df.columns:
        if numerical_df[col].nunique() < 10:
            plt.figure(figsize=(5,3))
            numerical_df[col].value_counts().plot(kind='bar')
            plt.title(col)
            plt.show()

def plot_continuous_vars(X):
    for col in numerical_df.columns:
        if numerical_df[col].nunique() >= 10:
            plt.figure(figsize=(5,3))
            plt.hist(numerical_df[col], bins=20)
            plt.title(col)
            plt.show()
```

```python
plot_discrete_vars(numerical_df)
```
