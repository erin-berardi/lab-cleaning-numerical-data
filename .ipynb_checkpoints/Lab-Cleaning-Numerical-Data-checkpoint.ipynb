```python
# Importing libraries
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing data
df_customer = pd.read_csv(r"C:\Users\joaoa\Desktop\Ironhack\Labs\lab-cleaning-numerical-data\files_for_lab\we_fn_use_c_marketing_customer_value_analysis.csv")
df_customer
```

```python
# Standardizing columns names
cols = []
for i in range(len(df_customer.columns)):
    cols.append(df_customer.columns[i].lower().replace(" ","_"))
df_customer.columns = cols
df_customer
```

```python
# Changing effective to date column to datetime format
df_customer["effective_to_date"] = pd.to_datetime(df_customer["effective_to_date"])
df_customer.dtypes
```

```python
# Creating a new dataframe using only the numerical data
df_numerical = df_customer.select_dtypes(include = np.number)
df_numerical
```

```python
# Defining a function that differentiates between continuous and discrete variables
def diff_variables(data_frame):
    global df_continuous
    global df_discrete
    
    df_continuous = df_numerical
    df_discrete = df_numerical
    
    for col in data_frame.columns.values:
        count = 1
        for val in data_frame.loc[:,col].values:
            if int(val) != val:
                df_discrete = df_discrete.drop([col], axis = 1)
                print(col, "is a continuous variable")
                break
            elif count == 9134:
                df_continuous = df_continuous.drop([col], axis = 1)
                print(col, "is a discrete variable")
            count += 1

diff_variables(df_numerical)     
```

```python
# Multicollinearity

# The variables have low absolute correlation values between them.
# Total claim amount has the most positive correlation with monthly premium auto.
# Total claim amount has the most negative correlation with income.

correlations_matrix = df_numerical.corr()
mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True #
fig, ax = plt.subplots(figsize=(20, 20))
ax = sns.heatmap(correlations_matrix, mask=mask, annot=False)
plt.show()
```

```python
# Defining a function that plots continuous and discrete variables

# All the graphs show a positive asymmetry. It means that the 50% smallest values are more frequent.

def plot_variables(data_frame):
    for col in data_frame.columns.values:
        count = 1
        for val in data_frame.loc[:,col].values:
            if int(val) != val:
                sns.distplot(data_frame[col])
                # sns.histplot(data = data_frame, x = col, binwidth = data_frame[col].max()/20)
                plt.show()
                break
            elif count == 9134:
                if data_frame[col].nunique() > 40:
                    sns.histplot(data = data_frame, x = col, binwidth = data_frame[col].max()/20)
                    plt.show()   
                else:
                    plot = sns.countplot(data_frame, x = col)
                    plt.xticks(rotation = 90)
                    plt.show()
            count += 1

plot_variables(df_numerical)   
```

```python
# Looking for outliers in the continuous variables
# I decided to keep them

sns.boxplot(x = df_numerical["customer_lifetime_value"], data = df_numerical)
plt.show()

sns.boxplot(x = df_numerical["total_claim_amount"], data = df_numerical)
plt.show()
```

```python
# Removing null values
df_numerical = df_numerical.dropna()
```
