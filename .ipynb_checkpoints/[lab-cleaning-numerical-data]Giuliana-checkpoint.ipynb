```python
# Rename the columns so they follow the PE8 (snake case) (lower_case_with_underscores)

cols = []

for a in range(len(customer_df.columns)):
    cols.append(customer_df.columns[a].lower().replace(' ', '_'))
    
customer_df.columns = cols

customer_df.columns
```

```python
# Change effective to date column to datetime format

customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
customer_df['effective_to_date']
```

```python
# Create a new dataframe using only the numerical data called numerical_df. Use this for the following questions

numerical_df = customer_df.select_dtypes(include=[np.number])
numerical_df
```

```python
# Define a function that differentiates between continuous and discrete variables 
# Hint: The number of unique values might be useful 
# Store continuous data into a continuous_df variable and do the same for discrete_df

print(numerical_df['customer_lifetime_value'].nunique())
print(numerical_df['income'].nunique())
print(numerical_df['monthly_premium_auto'].nunique())
print(numerical_df['months_since_last_claim'].nunique())
print(numerical_df['months_since_policy_inception'].nunique())
print(numerical_df['number_of_open_complaints'].nunique())
print(numerical_df['number_of_policies'].nunique())
print(numerical_df['total_claim_amount'].nunique())
numerical_df.dtypes
```

```python
# The threshold is randomn. in effect, I believe anything above a 15 threshold could be continuous, 
# but I went with 20 o be more conservative

def cont_or_disc (df):
    global continuous_df, discrete_df 

    continuous = []
    discrete = [] 

    for column in df.columns:
        unique_values = df[column].nunique()
        if unique_values >= 20:
            continuous.append(column)
        else:
            discrete.append(column)

    continuous_df = numerical_df[continuous]
    discrete_df = numerical_df[discrete]

    print("Continuous :", continuous_df.columns)
    print("Discrete :", discrete_df.columns)


cont_or_disc(numerical_df)



```

```python
# Create a function to plot every discrete variable
def discrete(df):
    for column in df.columns:
        plt.hist(data=df, x=column)
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()


discrete(discrete_df)       
```

```python

# Do the same with the continuous variables 
# Look for outliers in the continuous variables. (HINT: There’s a good plot to do that!)


def continuous(df):
    for column in df.columns:
        sns.boxplot(data=df, x=column)
        plt.title(f'Box Plot of {column}')
        plt.xlabel(column)
        plt.show()    

continuous(continuous_df)       
```
