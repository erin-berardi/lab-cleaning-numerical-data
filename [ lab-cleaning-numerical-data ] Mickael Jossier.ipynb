```python
# Did you find outliers? Comment what you will do with them. If you decide to remove them, do that now.
# will keep it for the discrete Data as they are not changing the target varibale and I will keep them for the continous Data as they are the main variable , creating the total claim amount

# anyway just showing the code to remove them in case 



# Create a copy of the original DataFrame to keep the original data
df_no_outliers = Numerical_df.copy()

# Loop through each column in the DataFrame
for column in Numerical_df.columns:
    Q1 = Numerical_df[column].quantile(0.25)
    Q3 = Numerical_df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Create a mask to identify outliers
    outlier_mask = (Numerical_df[column] < lower_bound) | (Numerical_df[column] > upper_bound)

    # Remove outliers from the column
    df_no_outliers[column] = Numerical_df[column][~outlier_mask]

df_no_outliers

```
